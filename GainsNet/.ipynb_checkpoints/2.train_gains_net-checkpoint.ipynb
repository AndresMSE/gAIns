```python
from gains_net import GainsNet
import numpy as np
from gains_plotter import GainsPlottter
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
import tensorflow as tf
# physical_devices = tf.config.list_physical_devices('GPU')
# tf.config.experimental.set_memory_growth(physical_devices[0], enable=True)
```

```python
npdata_path = './npdata_new'
# Load data
X_train = np.load(f"{npdata_path}/X_train.npy", allow_pickle=True)
y_train = np.load(f"{npdata_path}/y_train.npy", allow_pickle=True)
X_test = np.load(f"{npdata_path}/X_test.npy", allow_pickle=True)
y_test = np.load(f"{npdata_path}/y_test.npy", allow_pickle=True)
yc_train = np.load(f"{npdata_path}/X_val.npy", allow_pickle=True)
yc_test = np.load(f"{npdata_path}/y_val.npy", allow_pickle=True)
```

```python
# Parameters
LR = 0.0001
BATCH_SIZE = 128
N_EPOCH = 50

input_dim = X_train.shape[1]
feature_size = X_train.shape[2]
output_dim = y_train.shape[1]
```

```python
gainsnet = GainsNet(X_train, y_train, X_test, y_test)
gainsnet_model = gainsnet.get_lstm_model(LR)
```

```python
history = gainsnet.fit_model(gainsnet_model, BATCH_SIZE, N_EPOCH)
```

```python
GainsPlottter.plot_train_results(history)
```

```python
yhat = gainsnet_model.predict(X_test, verbose=0)
yhat
```

```python
rmse = sqrt(mean_squared_error(y_test, yhat))
rmse
```
